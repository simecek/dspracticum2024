```python
import pandas as pd

csv_filepath = 'diabetes.csv'
df = pd.read_csv(csv_filepath)

df
```

## Some cleaning etc.

```python
mean_blood_pressure = df[df['BloodPressure'] != 0]['BloodPressure'].mean()
df['BloodPressure'] = df['BloodPressure'].replace(0, mean_blood_pressure)
```

```python
mean_blood_pressure = df[df['SkinThickness'] != 0]['SkinThickness'].mean()
df['SkinThickness'] = df['SkinThickness'].replace(0, mean_blood_pressure)
```

## Train test split

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
X=df.drop('Outcome',axis=1)
Y=df['Outcome']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print(X_train.shape,X_test.shape)
```

## Train and test Random Forest classifier

```python
model = RandomForestClassifier(n_estimators=10)
model = model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
```
